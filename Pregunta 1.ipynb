```python
#MEDIA
import math
import numpy as np
import pandas as pn
print('Primer Data SET')
A=pn.read_csv('PoliceFatalities2.csv',skiprows=0,usecols=[0,2,7,8,9,10,11])
V1=np.array(A)

c=0
col=0
prom=0
cc=0
sum1=0
des1=0
print('Medias por Columnas seleccionadas')
while col<7:
    prom=0
    c=0
    while c<len(V1):
        prom=prom+V1[c,col]
        c+=1
    promedio=prom/len(V1)
    print('promedio',col)
    print(promedio)
    
    #de paso la desviacion estandar
    while cc<len(V1):
        sum1=sum1+(V1[cc,col]-promedio)**2
        cc+=1
    sum1=sum1/len(V1)
    print('Desviacion estandar col: ',col)
    print(math.sqrt(sum1))
    promedio=0
    sum1=0
    cc=0
    col=col+1
        

print('Segundo Data SET')
B=pn.read_csv('room-occupancy.csv',skiprows=0)
V2=np.array(B)
c=0
col=0
prom=0
print('Medias para todas las columnas')
while col<6:
    prom=0
    c=0
    while c<len(V2):
        prom=prom+V2[c,col]
        c+=1
    promedio=prom/len(V2)
    print('promedio',col)
    print(promedio)
    #de paso la desviacion estandar
    while cc<len(V2):
        sum1=sum1+(V2[cc,col]-promedio)**2
        cc+=1
    sum1=sum1/len(V2)
    print('Desviacion estandar col: ',col)
    print(math.sqrt(sum1))
    promedio=0
    sum1=0
    cc=0
    col=col+1
```

```python
#FUNCION PARA MODA
def moda(dataset):
    frequency = {}

    for value in dataset:
        frequency[value] = frequency.get(value, 0) + 1

    most_frequent = max(frequency.values())

    modas = [key for key, value in frequency.items()
                      if value == most_frequent]

    return modas
print(moda(V1))

#MODA
oned = np.array(V1)
oned2=np.transpose(oned)

lista1=oned2.tolist()
print('Primer Data SET')
print('moda por columna')
c2=0
while c2<7:
    print(moda(lista1[c2]))
    c2+=1
onedd = np.array(V2)
onedd2=np.transpose(onedd)

lista2=onedd2.tolist()
print('SEGUNDO Data SET')
print('moda por columna')
c2=0
while c2<6:
    print(moda(lista2[c2]))
    c2+=1
```
